# 2.2 Split Data
70% - Training
20% - Validation
10% - Testing

```python
#len(data) # number of batches
train_size = int(.7*(len(data)))
val_size   = int(.2*(len(data)))
test_size  = int(.1*(len(data))) + 1
```

```python
#double check all batches are accounted for 
train_size + val_size + test_size == len(data)
```

```python
# Use 'take' and 'skip' to populate data
train = data.take(train_size)
val   = data.skip(train_size).take(val_size)
test  = data.skip(train_size + val_size).take(test_size)
```

# 3.1 Build Deep Learning Model

```python
from tensorflow.keras.models import Sequential #Simple, can use Functional also
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Dense, Flatten, Dropout #why these packages - 50 minute mark in video
```

```python
# Create architectral for model
model = Sequential()
```

```python
# Adding Convelutional, with 16 filters with (3,3) pixel size and and 1 pixel strde. relu activation(negative values go to zero, positive stay the same) 
model.add(Conv2D(16, (3,3), 1, activation='relu', input_shape=(256,256,3))) # first layer must be an input
model.add(MaxPooling2D())

model.add(Conv2D(32, (3,3), 1, activation='relu'))
model.add(MaxPooling2D())
model.add(Conv2D(16, (3,3), 1, activation='relu'))
model.add(MaxPooling2D())
model.add(Flatten())
model.add(Dense(256, activation='relu'))
model.add(Dense(1, activation='sigmoid'))
```

```python
model.compile('adam', loss=tf.losses.BinaryCrossentropy(), metrics=['accuracy'])
```

```python
model.summary()
```

# Train

```python
logdir = 'Logs'
tensorboard_callback = tf.keras.callbacks.TensorBoard(log_dir=logdir)
```

```python
hist = model.fit(train, epochs=10, validation_data=val, callbacks=[tensorboard_callback])
```
